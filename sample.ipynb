- $ EX = {\lambda}^{-1} $
- $ EX^2 = {\lambda}^{-2} $
- медиана = $ ln(2)/\lambda $

### Данные по выборке


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lm = 10
def F(x):
    return 0 if x < 0 else 1 - np.e ** (-lm * x)

def generate_sample(sample_size):
    x = np.random.uniform(size=(sample_size,1))
    e = np.random.normal(0,0.01,size=x.shape)
    return x, [F(x0) for x0 in x]
```

```python
x, y = generate_sample(100)
plt.scatter(x,y)
```
